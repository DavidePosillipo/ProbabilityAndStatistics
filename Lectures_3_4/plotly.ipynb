```python
import plotly as py
```

```python
py.offline.init_notebook_mode(connected=True)
```

```python
# building blocks of PLotly
import plotly.grapht_objs as go 
```

```python
import numpy as np
x = np.linspace(0, 2*np.pi)
```

```python
# Trace 
trace0 = dict(type = "scatter", x = x, y = np.sin(x))
trace1 = dict(type = "scatter", x = x, y = np.cos(x))
```

```python
# COnfig the traces 
fig = go.Figure(data = [trace0, trace1])
```

```python
# Creating a layout 
my_layout = dict(title = "SIN and COS functions", 
             xaxis = dict(title = "x"), 
             yaxis = dict(title = "f(x)"))
```

```python
# Config the traces using the layout 
fig = go.Figure(data = [trace0, trace1], layout = my_layout)
```

```python
# in the browser
py.offline.plot(fig)
```

```python
# in the notebook 
py.offline.iplot(fig)
```
